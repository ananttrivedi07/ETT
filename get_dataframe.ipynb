Getting the data from CSV and making a dataframe for required variables

```python
import os
import glob
import pandas as pd
```

```python
# making the timestamps readable for input and lap.csv
def make_timestamps(file_with_timestamp):
    starting_timestamp = file_with_timestamp.loc[0, 'timestamp']
    file_with_timestamp['timestamp'] -= starting_timestamp - 1  
    return file_with_timestamp  
```

```python
save_drivers = []

# drivers that the user wants to test for
def input_drivers():
    save_drivers = []
    while len(save_drivers) <= 11:
        driver = input("Enter the driver name: ")
        if driver == "end":
            break
        else:
            save_drivers.append(driver)
    
    return iterate_drivers(save_drivers)
    

# iterate over all the drivers and get input and lap data for each and everyone of them
def iterate_drivers(save_drivers):
    for driver in save_drivers:
        return get_driver_data(driver)
    

def get_driver_data(driver_name):
    
    # array for the saving the files:
    data = [] # input.csv, all_laps
    # collect the data the specific driver:
    
    path = os.path.join(r"C:\Users\anant\OneDrive\Desktop\ETT\Green_team_selected_driver", driver_name)
    # getting all the csv files for a driver:
    csv_files = glob.glob(os.path.join(path, "*.csv"))
    
    
    # for all files for that driver, we are accessing only input and lap.csv
    for f in csv_files:
        
        file_name = f.split(os.path.sep)[-1]
        if file_name == "input.csv":
            # selecting only some of the variables from the file
            input_col_list = ["steer","timestamp"]
            car_input_info = pd.read_csv(f, usecols=input_col_list)
            # sending the file to fix the timestamps and creating a df
            df_input = pd.DataFrame(make_timestamps(car_input_info))
        elif file_name == "lap.csv":
            lap_col_list = ["lap count", "timestamp"]
            car_lap_info = pd.read_csv(f, usecols=lap_col_list)
            # sending the lap information to fix the timestamps.
            df_lap = pd.DataFrame(make_timestamps(car_lap_info))
    
    df_final = df_input.merge(df_lap, on="timestamp")

    return df_final
```
